#**1. Introduction**

The Stroke Prediction Dataset represents a valuable repository of health-related information, specifically designed to explore, and analyze the factors influencing the occurrence of strokes among a diverse set of individuals. Comprising 5110 collected observations, this dataset offers a insight into the various aspects of human health and lifestyle.

##**1.1. Data Description**

**Key Attributes of the Dataset:**

1.   **id:** Unique identifier for each patient.
2.   **gender:** Gender of the patient (male, female, or other).
3.   **age:** Age of the patient.
4.   **hypertension:** Whether the patient has hypertension (1 for yes, 0 for no).
5.   **heart_disease:** Whether the patient has heart disease (1 for yes, 0 for no).
6.   **ever_married:** Marital status of the patient.
7.   **work_type:** Type of work the patient is engaged in (e.g., private, self-employed, etc.).
8.   **Residence_type:** Type of residence (urban or rural).
9.   **avg_glucose_level:** Average glucose level in the patient's blood.
10.  **bmi:** Body mass index of the patient.
11.  **smoking_status:** Smoking status of the patient (e.g., never smoked, formerly smoked, or currently smoking).
12.  **stroke:** The target variable, indicating whether the patient has had a stroke (1 for yes, 0 for no).

##**1.2. Problem Statement**

Stroke is a severe medical condition that can lead to long-term disability and, in some cases, even death. Timely prediction of stroke risk can enable proactive healthcare interventions and, potentially, the prevention of stroke-related cases. Therefore, we aim to predict the occurrence of strokes in individuals based on their health-related attributes.

##**1.3. Objectives**

The primary objective of this analysis is to create a predictive model that can accurately identify individuals at risk of experiencing a stroke based on their health attributes. By doing so, we aim to achieve the following:



*    Reduce dimensionality of the dataset
*    Identify key predictors of Stroke Occurrence

PCA is a dimensionality reduction technique that can help us identify the most influential attributes contributing to stroke occurrence. By reducing the dataset's dimensionality, we can focus on the principal components that explain the most variance in the data. This can reveal which attributes have the strongest impact on stroke risk.

##**5.1. Feature selection**

```python
from sklearn.feature_selection import SelectKBest, chi2

# Define the number of features to select
num_features_to_select = 5

# Select the top k features using chi-squared test
X = data_encoded.drop(columns=['stroke'])  # Features
y = data_encoded['stroke']  # Target variable

# Apply feature selection
best_features = SelectKBest(score_func=chi2, k=num_features_to_select)
X_new = best_features.fit_transform(X, y)

# Get the indices of selected features
selected_feature_indices = best_features.get_support(indices=True)

# Get the names of selected features
selected_features = X.columns[selected_feature_indices]

print("Selected Features:")
print(selected_features)
```

##**5.4.PCA**

```python
#Calculate the covariance matrix
cov_matrix = np.cov(X_standardized, rowvar=False)
cov_matrix
```

```python
#Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
#Identify Principal Components
#Sort eigenvalues in decreasing order and identify the top N components
sorted_indices = np.argsort(eigenvalues)[::-1]
top_N = 2  # Specify the number of components to retain
eigenvalues = eigenvalues[sorted_indices][:top_N]
eigenvectors = eigenvectors[:, sorted_indices][:, :top_N]
```

```python
#Reduce Dimensions
pca = PCA(n_components=top_N)
data_reduced = pca.fit_transform(X_standardized)
data_reduced
```

##5.5. **PCA results**

```python
# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Plot explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Ratio vs. Number of Principal Components')
plt.show()
```
